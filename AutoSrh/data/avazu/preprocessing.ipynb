```python
train_df_le_30 = train_df_le_10.groupby('device_id').filter(lambda x: len(x)>=30)
```

```python
train_df_se_3e7 = train_df_le_30.groupby('device_id').filter(lambda x: len(x)<3e7)
```

```python
train_df_se_3e7.shape
```

```python
train_df_se_3e7.groupby('device_id').count()['id'].describe()
```

```python
train_df_se_3e7[train_df_se_3e7['click']==0].shape
```

```python
train_df_se_3e7 = train_df_se_3e7.rename(columns={'device_id': 'uid', 'id': 'iid'})
```

```python
train_df_se_3e7.columns
```

```python
new_columns_order  = ['click', 'uid', 'C1', 'banner_pos', 'site_id', 'site_domain',
       'site_category', 'app_id', 'app_domain', 'app_category', 'device_model', 'device_type', 'device_conn_type', 'C14',
       'C15', 'C16', 'C17', 'C18', 'C19', 'C20', 'C21']
train_df_se_3e7_newOrder = train_df_se_3e7[new_columns_order]
```

```python
train_df_se_3e7_newOrder.head()
```

```python
last_max=-1
for col in train_df_se_3e7_newOrder.columns.values[1:]:
    train_df_se_3e7_newOrder[col]=train_df_se_3e7_newOrder[col].astype("category").cat.codes
    train_df_se_3e7_newOrder[col]+=last_max+1
    last_max = train_df_se_3e7_newOrder[col].max()
```

```python
train_df_se_3e7_newOrder.describe()
```

```python
train_df_se_3e7_newOrder=train_df_se_3e7_newOrder.rename(columns={'click': 'is_click'})
```

```python
id_cols = train_df_se_3e7_newOrder.columns.values[1:]
for col in id_cols:
    train_df_se_3e7_newOrder[col+'_val']=pd.Series(1, train_df_se_3e7_newOrder[col].index)
```

```python
train_df_se_3e7_newOrder.head()
```

```python
train_df_se_3e7_newOrder.to_csv('avazu.csv', index=False)
```
