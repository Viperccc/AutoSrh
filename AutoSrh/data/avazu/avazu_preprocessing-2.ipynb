```python
from sklearn.model_selection import train_test_split
import pickle
import os


data  = np.insert(idd, 0, values=label, axis=1)
print(data)
```

```python
from sklearn.model_selection import train_test_split
import pickle
import os


def train_val_test_split(data_df, test_ratio=0.1, val_ratio=0.2, seed=12):
    train, test = train_test_split(data_df, test_size=test_ratio, random_state=seed)
    train, val = train_test_split(train, test_size=val_ratio / (1 - test_ratio), random_state=seed)
    return train, val, test

print("Splitting...")
train_data, val_data, test_data = train_val_test_split(data, test_ratio=0.1, val_ratio=0.2, seed=12)

print("Splitting finish")



```

```python
print(train_data)
```

```python
train_x = train_data[:,1:]    
train_y = train_data[:,0]
val_x = val_data[:,1:]    
val_y = val_data[:,0]
test_x = test_data[:,1:]    
test_y = test_data[:,0]
```

```python
import torch
from torch.utils.data import Dataset

class Avazu_Dataset(Dataset):
    def __init__(self, train_x,train_y):
        self.x = torch.LongTensor(train_x)
        self.y = torch.Tensor(train_y)

    def __getitem__(self, idx):
        return self.x[idx], torch.ones_like(self.x[idx], dtype=torch.float32), self.y[idx]

    def __len__(self):
        return len(self.x)
```

```python
print(idd.max())
num_features=idd.max()+1
```

```python
print("The dataset has been preprocessed.")
result = [Avazu_Dataset(train_x,train_y), Avazu_Dataset(val_x,val_y), Avazu_Dataset(test_x,test_y), num_features]
```

```python
with open(f"click.pickle", 'wb') as handle:
    pickle.dump(result, handle,protocol = 4)
    print("File saved.")
```
